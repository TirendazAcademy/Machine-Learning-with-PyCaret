# Pycaret with DagsHub

This notebook shows how Pycaret will work with DagsHub.

```python
# Installing the necessary libraries.
!pip install --pre pycaret
!pip install mlflow dagshub
```

```python
# Importing pycaret
import pycaret
pycaret.__version__
```

```python
# Loading the dataset
from pycaret.datasets import get_data
data = get_data('insurance')
```

```python
# Connecting dagshub
from pycaret.regression import *
s = setup(data, target = 'charges', 
          session_id = 123, 
          log_experiment = 'dagshub',
          experiment_name = 'project_insurance',
          log_data = True)
```

```python
# Finding the best model
best = compare_models()
```

```python
# Saving the best model
save_model(best, 'best_insurance_model')
```
